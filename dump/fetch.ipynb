```python
# Daily
import csv
import requests
import json

config = {}
with open('env.json') as config_file:
    config = json.load(config_file)

company = "tatapower"
url = "https://www.alphavantage.co//query?function=TIME_SERIES_DAILY&symbol="+company+".bse&apikey="+config.get("API_KEY")

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)


response= response.json()

time_series = response["Time Series (Daily)"]

csv_file_name = 'time_series.csv'

header = ["date"] + list(time_series[next(iter(time_series))].keys())

with open(csv_file_name, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(header)
    
    for date, values in time_series.items():
        row = [date] + list(values.values())
        writer.writerow(row)

print(f"Data has been written to {csv_file_name}")


```
